```python
# Importer les bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Générer des données synthétiques à des fins d'illustration
X, _ = make_blobs(n_samples=300, centers=6, random_state=0, cluster_std=0.60)

# Créer un modèle de clustering hiérarchique avec 2 clusters
agglomerative_clustering = AgglomerativeClustering(n_clusters=2)

# Adapter le modèle aux données et prédire les clusters
labels = agglomerative_clustering.fit_predict(X)

# Afficher les données et les clusters sur la même figure
plt.figure(figsize=(10, 5))

# Afficher les données initiales
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Composant 1')
plt.ylabel('Composant 2')
plt.title('Data Points')

# Afficher les clusters obtenus
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.xlabel('Composant 1')
plt.ylabel('Composant 2')
plt.title('Hierarchical Clustering')

plt.tight_layout()
plt.show()
```
