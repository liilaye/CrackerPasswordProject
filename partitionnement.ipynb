```python
# Importer les bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Générer des données synthétiques à des fins d'illustration
X, _ = make_blobs(n_samples=300, centers=6, random_state=0, cluster_std=0.60)
# Créer un modèle K-means avec 2 clusters
kmeans = KMeans(n_clusters=2)

# Adapter le modèle aux données
kmeans.fit(X)

# Obtenir les centres de cluster et les étiquettes des clusters pour chaque point de données
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Afficher les données et les clusters sur la même figure
plt.figure(figsize=(10, 5))

# Afficher les données initiales
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data Points')

# Afficher les clusters obtenus
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')

plt.tight_layout()
plt.show()
```
